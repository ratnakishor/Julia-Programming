# Numbers in Julia

## Integers
* Whole-valued positive or negative number or 0.
* Default integer type depends on the system architecture
* Sys.WORD_SIZE  variable represent the system word size

```julia
Sys.WORD_SIZE
```

```julia
num1 = 89
println(num1)
println(typeof(num1))
```

```julia
num2 = -94
println(num2)
println(typeof(num2))
```

### typemin()
* Returns the minimum value that a numeric type can represent.

### typemax()
* Return the maximum value that a numeric type can represent.

```julia
println(typemin(num1))
println(typemax(num1))
```

```julia
num3 = UInt8(num1)
println(num3)
println(typeof(num3))
println(typemin(num3))
print(typemax(num3))
```

### Numbers with different number systems.

```julia
num4 = 0xa
num5 = 0b1010
num6 = 0o12
println(num4)
println(num5)
println(num6)
```

## Floating-point Numbers

* Numbers that contain floating decimal point
* Can be represented using Standard form or Scientific notation (using e) 

```julia
f1 = 13.7
println(typeof(f1))
```

```julia
f2 = -1.0
println(typeof(f2))
```

```julia
f3 = 2e2
println(f3)
println(typeof(f3))
```

* Float32 literals can be represented using f in the place of e

```julia
f5 = 23f-2
println(f5)
println(typeof(f5))
```

* Float64 to Float32 conversion

```julia
f64 = 1.0
println(typeof(f64))

f32 = Float32(f64)
println(typeof(f32))
```

* Julia allows underscore( _ ) as a digit separator

```julia
val1 = 1_000
print(val1)
```

## Complex Numbers

* Numbers in the form a+bim

   --> a and b are real numbers represents real and imaginary parts respectively
   
   --> im = squrt(-1)

```julia
c1 = 2 + 4im
println(c1)
println(typeof(c1))
```

```julia
c2 = -2.9 + 4im
println(c2)
println(typeof(c2))
```

* real() function returns the real part
* imag() function returns the imaginary part

```julia
c3 = 20-4.5im
println(real(c3))
println(imag(c3))
```

# Rational Numbers

* Numbers expressed as the fraction of two integers, p/q.
* Operator used in Julia is //
* numerator(): Function to extract numerator
* denominator(): Function to extract denominator


```julia
r1 = 4//5
println(r1)
println(typeof(r1))
println(numerator(r1))
println(denominator(r1))
```

* Construction of infinite Rational numbers is acceptable

```julia
12//0
```

* float() function can be used to convert the complex number as floating-point number

```julia
r2 = 23//87
f1 = float(r2)
println(f1)
println(typeof(f1))
```
