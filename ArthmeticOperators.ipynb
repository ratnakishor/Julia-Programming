<h1 style = "color:Brown"> Arithmetic Operators in Julia</h1>

<h3 style = "color:Orange">Binary Operators</h3>

#### Variables a and b

```julia
# Variables 
a = 15
b = 4;
```

#### Binary Addition

```julia
println("a + b = ", a + b)
```

#### Binary Subtraction

```julia
println("a - b = ", a - b)
println("b - a = ", b - a)
```

#### Binary Multiplication

```julia
println("a * b = ", a * b)
```

#### Binary Division

```julia
println("a / b = ", a / b)
println("a ÷ b = ", a ÷ b)
println("b / a = ", a \ b)
println("a % b = ", a % b)
println("a % b = ", rem(a, b))
```

#### Exponentiation

```julia
println("a ^ b = ", a ^ b)
```

<h3 style = "color:Orange">Unary Operators</h3>

```julia
println(+a)
println(-a)
```

<h3 style = "color:Orange">Arithmetic Operations on Other Data Types</h3> 

#### Floating-point Numbers

```julia
x = 12.5
y = 3.45

println("x + y = ", x + y)
println("x - y = ", x - y)
println("x * y = ", x * y)
println("x / y = ", x / y)
```

#### Complex Numbers

```julia
k = 2 + 3im
l = 4 + 6im

println("k + l = ", k + l)
println("k - l = ", k - l)
println("k * l = ", k * l)
println("k / l = ", k / l)
```

#### Rational Numbers

```julia
p = 1//2
q = 3//5

println("p + q = ", p + q)
println("p - q = ", p - q)
println("p * q = ", p * q)
println("p / q = ", p / q)
```

<h3 style = "color:Orange">Alternate Notation</h3>

#### Operator(operand1, operand 2, ......, operand n)

```julia
n1 = +(4, 5, 3, 2)
println(n1)

n2 = -(4, 7)
println(n2)

n3 = *(3, 4, 2)
println(n3)

n4 = /(34, 4)
println(n4)
```

<h3 style = "color:Orange">Numeric Literal Coefficients</h3>


* Julia allows a variable to  be preceeded by a numeric literal to imply multiplication


* 2\*x or 2x or x*2

```julia
x = 2

y = 2*x + 2
println(y)

y1 = 2x + 2
println(y1)

y2 = 2x^2 + 2
println(y2)
```

#### Roots of the Quadratic Equation


```julia
# Solving Quadratic equation
a = 5
b = 6
c = 1
c1 = (-b - sqrt(b^2 - 4a*c)) / 2a
c2 = (-b + sqrt(b^2 - 4a*c)) / 2a
println(c1)
println(c2)
```
