<h1 style = "color:Red">Numeric Comparison Operators</h1>


* To compare two values


* Results in Boolean value(true or false).



```julia
x = 100
y = 50

println("x Equals to y: ", x == y)

println("x Not Equals to y: ", x != y)

println("x Less than y: ", x < y)

println("x Less than or qual to y: ", x <= y)

println("x Greater than y: ", x > y)

println("x Greater than or equal to y: ", x >= y)
```

<h1 style = "color:Red">Logical Operators</h1>


* Used to make decisions based on multiple conditions 




```julia
a = 5
b = 4
c = -3

println(!(a > b))

println( (a > b) && (b > c ))

println( (a > b) || (b > c ))
```

<h1 style = "color:Red">Ascertain Functions</h1>

1. **isa(variable, type)**: Tests if a variable of a given type

```julia
println(isa(56, Int64))

println(isa(56, Int32))
```

2. **isnumeric(character)**: Tests whether a character is numeric

```julia
println(isnumeric('a'))

println(isnumeric('8'))
```

3. **iseven(Integer)**: Tests whether the given integer is even

```julia
println(iseven(6))

println(iseven(5))
```

4. **isodd(Integer)**: Tests whether the given integer is odd

```julia
println(isodd(6))

println(isodd(5))
```

5. **ispow2(number)**: Tests whether the given number is an ineger power of 2

```julia
println(ispow2(8))

println(ispow2(9))
```

6. **isfinite(number)**: Tests whether the given number is finite

```julia
println(isfinite(100))

println(isfinite(19/0))

println(isfinite(Inf))
```

7. **isinf(number)**: Tests whether the given number is infinite

```julia
println(isinf(100))

println(isinf(19/0))

println(isinf(Inf))
```

8. **isnan(value)**: Tests whether the given value is NaN

```julia
println(isnan(9))

println(isnan(NaN))
```

9. **iszero(number)**: Tests whether the given number is zero

```julia
println(iszero(10))

println(iszero(0))
```

<h1 style = "color:Red">Ternary Operator</h1>


<h3>condition ? expression_A : expression_B</h3>

```julia
# Even or Odd

num = 501

iseven(num) ? println("Even number") : println("Odd number")
```
